# 0.2 Colab、GPU 與資料路徑設定

這份 Notebook 示範 TensorFlow 專案開始前常做的環境檢查：版本、硬體裝置、工作目錄、資料資料夾與簡單檔案讀寫。

## 1. 載入套件

先載入環境檢查與路徑處理會用到的套件。`pathlib.Path` 是後續 cookbook 建議使用的路徑管理方式。

```python
import os
import platform
from pathlib import Path

import numpy as np
import pandas as pd
import tensorflow as tf

print('Python:', platform.python_version())
print('TensorFlow:', tf.__version__)
```

## 2. 檢查 CPU / GPU 裝置

在 Colab 中若有開啟 GPU，`tf.config.list_physical_devices('GPU')` 會列出可用 GPU。若清單為空，通常代表目前 runtime 使用 CPU。

```python
cpu_devices = tf.config.list_physical_devices('CPU')
gpu_devices = tf.config.list_physical_devices('GPU')

print('CPU devices:', cpu_devices)
print('GPU devices:', gpu_devices)

if gpu_devices:
    print('目前偵測到 GPU，可以用於較大的神經網路訓練。')
else:
    print('目前未偵測到 GPU；小型範例仍可使用 CPU 執行。')
```

## 3. 檢查目前工作目錄

Notebook 在不同環境執行時，工作目錄可能不同。先印出目前位置，可以避免資料路徑設定錯誤。

```python
PROJECT_DIR = Path.cwd()
DATA_DIR = PROJECT_DIR / 'data'
OUTPUT_DIR = PROJECT_DIR / 'outputs'
MODEL_DIR = OUTPUT_DIR / 'models'
FIGURE_DIR = OUTPUT_DIR / 'figures'

for directory in [DATA_DIR, OUTPUT_DIR, MODEL_DIR, FIGURE_DIR]:
    directory.mkdir(parents=True, exist_ok=True)

print('PROJECT_DIR:', PROJECT_DIR)
print('DATA_DIR:', DATA_DIR)
print('OUTPUT_DIR:', OUTPUT_DIR)
```

## 4. 建立一份示範資料

這裡建立一份小型 CSV，模擬實務上把原始資料放在 `data/` 底下，再用 Pandas 載入。

```python
demo_df = pd.DataFrame({
    'feature_1': [1.2, 2.4, 3.1, 4.8],
    'feature_2': [10, 12, 13, 18],
    'target': [0, 0, 1, 1]
})

csv_path = DATA_DIR / 'demo_dataset.csv'
demo_df.to_csv(csv_path, index=False)

loaded_df = pd.read_csv(csv_path)
print('saved to:', csv_path)
loaded_df
```

## 5. 使用相對於專案根目錄的路徑

後續程式盡量不要把資料路徑寫成固定絕對路徑。使用 `PROJECT_DIR / 'data' / 'file.csv'` 這種寫法，會更容易在 Colab、本機與 GitHub 專案之間切換。

```python
def describe_path(path: Path):
    return {
        'path': str(path),
        'exists': path.exists(),
        'is_file': path.is_file(),
        'suffix': path.suffix,
        'size_bytes': path.stat().st_size if path.exists() else None,
    }

pd.DataFrame([describe_path(csv_path)])
```
